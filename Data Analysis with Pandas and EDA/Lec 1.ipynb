### Describe method
- Statistical summary
 **Syntax**

```python
df.describe()
```

```python
df.describe(include='all')
```

```python
df.describe(exclude='object').round(2) ## name and city are object so excluded
```

```python
df.describe(percentiles=[.20,.30,.45,.87]).round(2)
```

```python
mystats=df.describe(percentiles=[.20,.30,.45,.87]).round(2)
print(mystats)
```
