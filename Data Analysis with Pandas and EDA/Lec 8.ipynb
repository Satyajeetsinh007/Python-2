```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris ## Library for ML
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
print(df)
df['target']=iris.target
df['species']=df['target'].map({
    0:'setosa',
    1:'versicolor',
    2:'verginica'
})
df.head(1)
```

```python
df.shape
```

## pd.get_dummies
- This method converts categorical columns to boolean(numerical columns)
- The purpose is to facilitate ML Analysis

```python
pd.get_dummies(df,columns=['species'],drop_first=True)
```

```python
df
```

# `np.where()`
- Condition based filtering of datasets

```python
df.head(1)
```

```python
df['size']=np.where(df["sepal length (cm)"]>6.5,'Large',np.where(df["sepal length (cm)"]>5.5,'Medium','Small'))
df.head()
```
