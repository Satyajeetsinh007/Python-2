### Q311) Desicion Tree For Diabetes 

```python
#  Diabetes Prediction using Random Forest
# You are given a dataset diabetes.csv containing medical records of patients.
# The dataset includes predictor variables such as BMI, Age, Insulin, Pregnancies, etc., and a target variable named Outcome.
# Your task is to:
# Load the dataset using pandas.
# Separate independent and dependent variables.
# Split dataset into training and testing sets.
# Train a Random Forest Classifier.
# Predict patient outcomes.
# Evaluate performance using:
# Confusion Matrix
# Accuracy Score
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/patelmanishv/Sem4Data/refs/heads/main/Datasets/diabetes.csv')
df.head()
```

```python
x=df.drop(columns='Outcome');
y=df['Outcome'];
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2);
```

```python
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=3);
```

```python
dtree.fit(x_train,y_train)
```

```python
prediction=dtree.predict([[1,180,90,20,0,30,0.245,45]])
print(prediction)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,classification_report
```

```python
cm=confusion_matrix(y_test,dtree.predict(x_test))
print("Confusion Matrix -\n",cm)
```

```python
print(accuracy_score(y_test,dtree.predict(x_test)))
```
