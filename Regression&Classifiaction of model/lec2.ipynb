##  A Template program for solving linear regression(FOR EXAM)

### 1. Import the library

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
```

### 2. Load the dataset

```python
iris=datasets.load_iris()
x=iris.data[:,2].reshape(-1,1) #Petal length
y=iris.data[:,3]# Petal width
print(x.shape)
print(y.shape)
```

```python
# print(x)
```

### Split the dataset

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### 4.Create table

```python
model=LinearRegression()
```

### 5.Fit the model(Training data will be send here)

```python
model.fit(x_train,y_train)
```

### print slope and intercept

```python
slope=model.coef_[0]
intercept=model.intercept_
print(f"The slope is {slope:.2f} and intercept is {intercept:.2f}")
```

### 6. Regression Line Equation


```python
print(f"The Linear regression equation is y={intercept:.2f}+{slope:.2f}*x")
```

```python
# for x=90 y=36.83
prediction=model.predict([[90]])
print(prediction)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
```

#### Evaluation

```python
mse=mean_squared_error(x_test,y_pred)
r2=r2_score(y_test,y_pred)
print(f"mean squared error is {mse:.2f} and R2 score is {r2:.2f}")
```

### Plot the regression line(Optional)

```python
plt.figure(figsize=(7,5))
plt.scatter(x_test,y_test,color='black',label='Actual Data')
plt.plot(x_test,y_pred,color='blue',label="RegressionLine",linewidth=2)
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.title("LINEAR Regression: PetalLength vs PetalWidth")
plt.legend()
x_min,x_max=plt.xlim()
y_min,y_max=plt.ylim()
equation=f"y={intercept:.2f}+{slope:.2f}*x"
plt.text(
    x_max-0.5*(x_max-x_min),
    y_max+0.5*(y_max-y_min),
    equation,fontsize=10,color='darkgreen',ha='right',
    bbox=dict(facecolor='white',alpha=0.8)
)
plt.show()
```
