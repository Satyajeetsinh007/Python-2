## Multiple Linear Regression
- in MLR there are multiple features or independent variables and one dependent varaible(target)

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
data=pd.DataFrame({
    'size':[2104,1600,2400,1416,3000],
    'year':[2003,1977,2001,1995,2000],
    'price':[399900,329900,369000,232000,539900]
})
```

```python
# Features and target
x=data[['size','year']];
y=data['price'];
print(x.shape);
print(y.shape);
```

```python
model=LinearRegression();
model.fit(x,y)
```

```python
slope=model.coef_[0];
intercept=model.intercept_;
print("slope=",slope," intercept=",intercept);
```

```python
new_input=pd.DataFrame({
    'size':[750],
    'year':[2009]
})
prediction=model.predict(new_input);
print("prediction of 750 sq area in 2009=",prediction[0])
```

## Polynomial Regression

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import seaborn as sns

iris=sns.load_dataset('iris');
# iris.head
# iris.shape;
# Define features and target
print(iris)
x=iris[['sepal_length']];
y=iris['petal_length'];
print(x.shape);
print(y.shape);
```

```python
# Create Polynomial features

poly=PolynomialFeatures(degree=2);
x_poly=poly.fit_transform(x);
model=LinearRegression();
model.fit(x_poly,y);
print(model.coef_);
print(model.intercept_)
```

```python
x_test_value=np.linspace(4,8,100).reshape(-1,1);
x_test=pd.DataFrame(x_test_value,columns=['sepal_length']);
x_test_poly=poly.transform(x_test);
y_pred=model.predict(x_test_poly);
print(y_pred)
```
