### Decision Tree Python implementation of job offer dataset

#### Import Libraries


```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/patelmanishv/Sem4Data/refs/heads/main/Datasets/job_offers.csv")
df.head(2)
```

```python
## Define Features;

x=df.drop(columns=['JobOffered']);
y=df['JobOffered'];
print(x.shape);
print(y.shape);
```

### Create a model


```python
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=3)
```

```python
dtree.fit(x,y)
```

```python
### Prediction
prediction=dtree.predict(x)
print(prediction)
```

### Plot the tree (not in syllabus)

```python
plt.figure(figsize=(12,8))
plot_tree(dtree,filled=True,feature_names=['CGPA','Communication','Aptitude','Programming Skill'],class_names=['No','Yes'])
plt.title("Decision tree of kob offer dataset");
plt.show()
```

### Evaluation of Metrics in Classification models
- Confusion Metrics
- accuracy score
- precision score
- recall score
- f1 score
- classification report

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,classification_report
```

### Confusion Metrics

```python
cm=confusion_matrix(y,prediction)
print("Confusion Matrix -\n",cm)
```

```python
TN,FP,FN,TP=cm.ravel();
print("True Negatives",TN);
print("False Positives",FP);
print("False Negatives",FN);
print("True Positives",TP);
```

### Accuracy Score
- accuracyscore=(TP+TN)/(TN+FP+FN+TP)

```python
accuracyscore=(TP+TN)/(TN+FP+FN+TP)
print('Accuracy score is ',accuracyscore)
print('Error rate is ',1-accuracyscore)
```

### Precision

```python
precisionscore=TP/(TP+FP);
print("Precision score is",precisionscore)
print(precision_score(y,prediction))
```

### Recall

```python
recallscore=TP/(TP+FN);
print("Recall score is",recallscore)
print(recall_score(y,prediction))
```

### f1 score 
- f1score=2*(prediction*recallscore)/(precisionscore+recallscore)

```python
f1score=2*(precisionscore*recallscore)/(precisionscore+recallscore)
print("F1 score",f1score)
print(f1_score(y,prediction))
```
